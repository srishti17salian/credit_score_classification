#### B) Replace missing values with appropriate values(4 Marks)

```python
# filling num values with median values
for i in num_cols:
    df.loc[df[i].isnull(),i] = df[i].median()

# filling cat values with mode values
for x in cat_cols:
    df.loc[df[x].isnull(),x] = df[x].mode()[0]
```

```python
# Distribution plot for num columns
t = 1
plt.figure(figsize=(13,20))
for i in cat_cols:
    if i !='Customer ID':
        plt.subplot(6,2,t)
        df[i].value_counts().plot(kind='bar')
        plt.title(i)
        t+=1
    
plt.tight_layout()
plt.show()
```

```python
# B. i)
def dunc(x):
    if 26>x>0:
        return('Young')
    elif 51>x>25:
        return('Middle Age')
    else:
        return ('Old')
```

```python
# ii)
df['Age_Groups'] = df.Age.apply(dunc)
```

```python
df.columns
```

```python
# iii)
plt.figure(figsize=(12,7))
df.groupby('Age_Groups')['Income (USD)'].mean().sort_values(ascending=False).plot(kind='bar')
plt.show()
```
